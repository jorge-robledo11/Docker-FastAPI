```python
import requests
import pandas as pd
import numpy as np
import asyncio
import aiohttp
import nest_asyncio

# Aplica nest_asyncio para que funcione en el entorno Jupyter
nest_asyncio.apply()
```

```python
data = pd.read_parquet('data/test_data.parquet')
data.head()
```

```python
np.random.seed(77)

size = 100
rows = np.random.randint(0, data.shape[0], size=size)
datapoints = data.iloc[rows, 1:].to_dict(orient='records')
datapoints
```

```python
# URL del endpoint predict
url = "http://127.0.0.1:5000/predict"

async def send_request(session, data):
    # Simula un retraso aleatorio entre 0.5 y 3 segundos
    delay = np.random.uniform(0.5, 3)
    await asyncio.sleep(delay)
    
    try:
        # También podemos establecer un timeout para la solicitud (por ejemplo, 10 segundos)
        async with session.post(url, json=data, timeout=10) as response:
            # Intenta decodificar la respuesta JSON
            return await response.json()
    except Exception as e:
        # En caso de error, se puede retornar un diccionario con el error
        return {"error": str(e)}

async def main():
    async with aiohttp.ClientSession() as session:
        # Creamos una lista de tareas para enviar todas las solicitudes concurrentemente
        tasks = [send_request(session, peticion) for peticion in datapoints]
        responses = await asyncio.gather(*tasks)
        for res in responses:
            print("Respuesta solicitud:", res)

# Ejecuta la función main usando el event loop existente
loop = asyncio.get_event_loop()
loop.run_until_complete(main())
```
